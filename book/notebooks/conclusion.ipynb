# Conclusions

In this project we have looked at the breakdown of the rotating wave approximation of the Jaynes cummimgs hamiltonian. Our main goal was to find out when this approximation fails by looking at many aspects of our system and seeing how the RWA causes deviation from the full JC hamiltonian including counter rotating terms. This difference between the hamiltonians can cause the breakdown in primarily the high regime. 

We analysed how the eigenenergies of the eigenstates changed for our system when there was an increase in coupling and found that RWA induces a significant error in the eigenenergies, when the coupling charcterised by the parameter g is increased. For strong coupling, g beyond 0.5 there is a error of more than 10%. This shows that RWA is only valid for the weak coupling regime. We can see that the squeezing effect, although preserved also is of much lower magnitude with the RWA. In addition, we also made the observation that central levels seem to deviate significantly over detuning at higher g, detuning itself at lower coupling does not play much of a role in the breakdown. 

We also found that when the coupling becomes stronger (Arround g $\approx$ 0.5) for the wigner quasi-probability function of the ground state breaks down. For the full Hamiltonian it evolves into a "cat" state while the RWA ground state develops multiple peaks without fringes as $g$ increases, failing to capture the superposition. This illustrates the RWAâ€™s limitations at strong coupling regime across its eigenstates and their eigenvalues.

Introducing noise into the system results in a relatively quick breakdown of the RWA. External noise or photons do not preserve the total excitation number which is the core assumption behind the RWA. Only considering noise, this breakdown generally occurs significantly faster for the comparison of the average photon number in the resonator with respect to the excited-state probability in the two-level system. Increasing relaxation time $T_1$ does not increase the coupling strength $g$ at which the breakdown occurs for the average photon number. However, when considering the excited-state probability, increasing the relaxation time results in a much larger deviation between the models, but decreases as the coupling strength $g$ is increased. This results in a peak at a coupling strength $g$ of approximately 0.2 to 0.3. Implementing the quantum measurement back-action results in a positive shift in the RWA breakdown for either comparisons. Increasing the measurement rate for both observables, the average photon number and the excited-state probability, results in a decrease of the maximum deviation. However, this is realized faster when comparing the excited-state probabilities.

In conclusion, the assumption that total excitation number between resonator and transmon (TLS) breaks down in with stronger coupling as the interaction energy is more significant. Through exploring the deviation in eigenenergies and eigenfunctions with a package like QuTIP, this phenomenon can be emperically evaluated.
