# Conclusions

In this project we have looked at the breakdown of the rotating wave approximation of the Jaynes cummimgs hamiltonian. Our main goal was to find out when this approximation fails by looking at many aspects of our system and seeing how the RWA causes deviation from the full JC hamiltonian. 

We analysed how the eigenenergies of the eigenstates changed for our system when there was a increase in coupling and found that RWA induces a significant error in the eigenenergies, when the coupling constant g is increased beyond 0.5 there is a error of more than 10%. This shows that RWA is only valy valid for the weak coupling, 

We also found that when the coupling becomes stronger the approximation breaks down (Arround g $\approx$ 0.5) for the wigner function of the ground state. For the full Hamiltonian it evolves into a "cat" state while the RWA ground state develops multiple peaks without fringes as $g$ increases, failing to capture the superposition.
