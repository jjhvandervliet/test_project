# **Measurement of a Two-Level System Coupled to a Resonator**
## Realistic Dynamics, Measurement, and Breakdown of the Rotating-Wave Approximation

This notebook is dedicated to studying a two-level quantum system coupled to a resonator governed by the Jaynes-Cummings (JC) Model, including dissipation and quantum measurement back-action. The full JC Model and the rotation-wave approximation, denoted as RWA henceforth, JC Model will be compared in order to determine when the RWA breaks down. This notebook therefore serves as a versatile generalization for systems that are governed by the models in questions.

Such systems appear in:  

- Cavity QED  
- Circuit QED  
- Trapped ions  
- Quantum dots in photonic cavities

## System and Hamiltonians

The system consists of:

- A resonator (frequency $w_r$) with annihilation operator $a$  
- A two-level system (qubit) with Pauli operators $\sigma_x, \sigma_y, \sigma_z$  

**JC Hamiltonian with RWA:**

$H_{JC} = w_r a^\dagger a + \frac{w_q}{2} \sigma_z + g(a^\dagger \sigma^- + a \sigma^+)$

**JC Hamiltonian without RWA:**

$H_{Full} = w_r a^\dagger a + \frac{w_q}{2} \sigma_z + g(a + a^\dagger)(\sigma^+ + \sigma^-)$

- $g$ is the coupling strength.  
- $w_q$ is the qubit frequency.  
- $a^\dagger a$ counts photons in the resonator.  

**References:**  

1. E.T. Jaynes and F.W. Cummings, *Comparison of quantum and semiclassical radiation theories with application to the beam maser*, Proc. IEEE 51, 89 (1963).  
2. P. Forn-Díaz et al., *Ultrastrong coupling regimes of light-matter interaction*, Rev. Mod. Phys. 91, 025005 (2019).


## Lindblad Noise and Measurement

The **Lindblad master equation** governs the dynamics:

$\frac{d\rho}{dt} = -i[H, \rho] + \sum_k \left(c_k \rho c_k^\dagger - \frac{1}{2}\{c_k^\dagger c_k, \rho\}\right)$

**Collapse operators $c_k$:**

| Collapse operator | Physical meaning |
|------------------|----------------|
| $c_1 = \sqrt{\kappa (1+n_{th})} a$ | Photon loss from the resonator |
| $c_2 = \sqrt{\kappa n_{th}} a^\dagger$ | Thermal excitation of the resonator |
| $c_3 = \sqrt{\gamma} \sigma^-$ | Qubit relaxation |
| $c_4 = \sqrt{\gamma_\phi} \sigma_z$ | Qubit dephasing |
| $c_5 = \sqrt{\gamma_\text{meas,cav}} a^\dagger a$ | Measurement-induced decoherence of cavity photons |
| $c_6 = \sqrt{\gamma_\text{meas,qubit}} \sigma^+ \sigma^-$ | Measurement-induced decoherence of qubit |

- **Measured observables:**  
  - Cavity photons: $a^\dagger a$  
  - Qubit excitation probability: $\sigma^+\sigma^-$  

**References:**  

1. H. Carmichael, *Statistical Methods in Quantum Optics 1*, Ch. 4  
2. H. Wiseman and G. Milburn, *Quantum Measurement and Control*, Ch. 3–4  
3. K. Jacobs and D. Steck, *A Straightforward Introduction to Continuous Quantum Measurement*, Contemp. Phys. 47, 279 (2006)


## Quantifying RWA Breakdown

To assess when the RWA fails, one has to compute the absolute deviation $\Delta$ between observables obtained with the full model and the RWA model:

- **Cavity photons:** $\Delta_\text{photons}(t) = | \langle a^\dagger a \rangle_\text{JC}(t) - \langle a^\dagger a \rangle_\text{Full}(t) |$  
- **Qubit excitation:** $\Delta_\text{qubit}(t) = | \langle \sigma^+\sigma^- \rangle_\text{JC}(t) - \langle \sigma^+\sigma^- \rangle_\text{Full}(t) |$  

To provide a single scalar measure for each coupling strength $g$, the maximum deviation over time $M$ is introduced:

- **Maximum deviation (cavity photons):** $M_\text{photons}(g) = \max_t \Delta_\text{photons}(t)$  
- **Maximum deviation (qubit excitation):** $M_\text{qubit}(g) = \max_t \Delta_\text{qubit}(t)$  

These maximum deviations indicate the worst-case effect of counter-rotating terms for a given coupling. High values signal significant breakdown of the RWA.

```python
import numpy as np
import qutip as qt
import matplotlib.pyplot as plt

# System Parameters
wr = 1.0 * 2*np.pi   # resonator frequency
wq = 1.2 * 2*np.pi   # qubit frequency
N  = 10              # resonator Hilbert space dimension
tlist = np.linspace(0, 50, 400)  # longer timescale

# Operators
a  = qt.destroy(N)
sm = qt.sigmam()
sp = qt.sigmap()
sz = qt.sigmaz()

# Initial state: 1 photon in resonator, qubit in ground
psi0 = qt.tensor(qt.basis(N,1), qt.basis(2,0))

# Lindblad Noise Parameters
kappa = 0.05          # resonator decay
gamma = 0.03          # qubit relaxation
gamma_phi = 0.01      # qubit dephasing
nth = 0.2             # thermal occupation

# Measurement Rates
gamma_meas_cavity = 0.01
gamma_meas_qubit  = 0.01

# Operators to Measure
cavity_photon_op = qt.tensor(a.dag()*a, qt.qeye(2))  # photon number
qubit_exc_op     = qt.tensor(qt.qeye(N), sm*sp)      # qubit excitation probability

# Collapse Operators
c_ops = [
    np.sqrt(kappa*(1+nth)) * qt.tensor(a, qt.qeye(2)),
    np.sqrt(kappa*nth)     * qt.tensor(a.dag(), qt.qeye(2)),
    np.sqrt(gamma)         * qt.tensor(qt.qeye(N), sm),
    np.sqrt(gamma_phi)     * qt.tensor(qt.qeye(N), sz),
    np.sqrt(gamma_meas_cavity) * cavity_photon_op,
    np.sqrt(gamma_meas_qubit)  * qubit_exc_op
]

# Coupling Strengths
g_list = np.linspace(0.01, 1.0, 40)

# Containers for deviations
deviation_photons = np.zeros((len(g_list), len(tlist)))
deviation_qubit   = np.zeros((len(g_list), len(tlist)))

# Compute deviations
for i, g in enumerate(g_list):

    # RWA Hamiltonian
    H_RWA = wr * qt.tensor(a.dag()*a, qt.qeye(2)) \
         + 0.5 * wq * qt.tensor(qt.qeye(N), sz) \
         + g * (qt.tensor(a.dag(), sm) + qt.tensor(a, sp))

    # Full Hamiltonian
    H_Full = wr * qt.tensor(a.dag()*a, qt.qeye(2)) \
           + 0.5 * wq * qt.tensor(qt.qeye(N), sz) \
           + g * qt.tensor(a + a.dag(), sp + sm)

    # Solve master equations
    res_RWA = qt.mesolve(H_RWA, psi0, tlist, c_ops,
                         [cavity_photon_op, qubit_exc_op])
    res_Full = qt.mesolve(H_Full, psi0, tlist, c_ops,
                           [cavity_photon_op, qubit_exc_op])

    # Compute absolute deviations
    deviation_photons[i, :] = np.abs(res_RWA.expect[0] - res_Full.expect[0])
    deviation_qubit[i, :]   = np.abs(res_RWA.expect[1] - res_Full.expect[1])

# -------------------------
# Plot heatmaps
# -------------------------
fig, axes = plt.subplots(1,2, figsize=(16,6))

# Cavity photon deviation
im0 = axes[0].imshow(deviation_photons, aspect='auto', origin='lower',
           extent=[tlist[0], tlist[-1], g_list[0], g_list[-1]],
           cmap='inferno')
axes[0].set_xlabel('Time')
axes[0].set_ylabel('Coupling g')
axes[0].set_title('Deviation in Cavity Photon Number')
fig.colorbar(im0, ax=axes[0], label='|RWA - Full|')

# Qubit excitation deviation
im1 = axes[1].imshow(deviation_qubit, aspect='auto', origin='lower',
           extent=[tlist[0], tlist[-1], g_list[0], g_list[-1]],
           cmap='inferno')
axes[1].set_xlabel('Time')
axes[1].set_ylabel('Coupling g')
axes[1].set_title('Deviation in Qubit Excitation Probability')
fig.colorbar(im1, ax=axes[1], label='|RWA - Full|')

plt.suptitle('RWA Breakdown: Cavity and Qubit Observables with Measurement', fontsize=16)
plt.show()

# =====================================================
# Maximum deviation measure
# =====================================================

# Compute maximum deviation over time for each coupling strength
max_dev_photons = np.max(deviation_photons, axis=1)
max_dev_qubit   = np.max(deviation_qubit, axis=1)

# Plot maximum deviations as a function of coupling g
plt.figure(figsize=(10,6))
plt.plot(g_list, max_dev_photons, label='Cavity photon number')
plt.plot(g_list, max_dev_qubit, label='Qubit excitation probability')
plt.xlabel('Coupling strength g')
plt.ylabel('Maximum absolute deviation |RWA - Full|')
plt.title('RWA Breakdown vs Coupling Strength')
plt.legend()
plt.grid(True)
plt.show()
```

## Interpretation

- **Left heatmap:** deviation of the resonator photon number between RWA and full model.  
- **Right heatmap:** deviation of qubit excitation probability.
